# FONCTION RECURSIVE

```python
liste_enregistrement = []
def descendre_les_categories(url, category ):
    browser.get(url)
    html_page_listing = browser.page_source # permet de prendre le code html de la page 
    soup = BeautifulSoup(html_page_listing , 'lxml') # créer la soupe avec le parser lxml
    categories = soup.find_all('li'  , class_ = 'a-spacing-micro apb-browse-refinements-indent-2') 
    if categories == [] :
        liste_enregistrement.append([url,category])
    else :
        for categorie in categories : # on va balayer toutes les catégories trouvées sur cette page
            lien = 'https://www.amazon.com'+ categorie.a.get('href') # on recréé le lien web
            categorie = categorie.span.text.strip() # via la fonction strip on récupère le nom de la catégorie
            #sous_categorie = [category, categorie] # ce code fonctionne mais donne des listes nested
            #category=[category]
            sous_categorie = category + [categorie] # à ce stade categorie est un texte mais pas category
            descendre_les_categories(lien,sous_categorie)
```
